# Seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
data = pd.read_csv('dataset/Advertising.csv')
data = data.drop('Unnamed: 0',axis=1)
levels = ['TV','Sales']
Udata = data[levels]
Udata.head()
```

## jointplot

```python
#seaborn joint plotting
sns.jointplot(x=Udata.TV,y=Udata.Sales,data=Udata,kind='reg')
```

## sns.lmplot

```python
g = sns.lmplot(x="Sales", y="TV", data=Udata)
plt.title('TV Advertisement vs Sales')
plt.show()
```
