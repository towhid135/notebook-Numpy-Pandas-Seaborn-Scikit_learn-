```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
df = pd.DataFrame({'x':[0,1,2,3],'y':[4,7,7,8]})
df.head()
```

```python
#plotting with the df
plt.plot(df.x,df.y,color='green', marker='*',linestyle='dashed',linewidth=2, markersize=12,label='line 1')
#plt.plot(df.y,df.x,color='green', marker='*',linestyle='dashed',linewidth=2, markersize=12,label='line 1')
plt.xlabel('Values of x')
plt.ylabel('Values of y')
```

```python
plt.scatter(df.x,df.y, s = 5, c = 'r')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
style.use('ggplot')
plt.scatter(df.x,df.y,s=5,c='k')
```
